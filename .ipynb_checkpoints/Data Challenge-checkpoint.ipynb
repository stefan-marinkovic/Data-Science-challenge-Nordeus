### Number of retention_d1 by something 

Function for aggregation:

```python
def retention_d1_by(by_something,rot=0,fig_siz=(7,5)):
    df_pom=df_train.groupby(by_something).agg({'registrations':'sum','returned':'sum'}).reset_index()
    df_pom['retention_d1']=df_pom['returned']/df_pom['registrations']
    plt.figure(figsize=fig_siz)
    plt.bar(df_pom[by_something],df_pom['retention_d1'],tick_label=df_pom[by_something].unique())
    plt.xticks(rotation=rot)
    plt.xlabel(by_something)
    plt.ylabel('retention_d1')
```

### Registration Type

```python
retention_d1_by('registration_type')
```

- 1 = Fresh Registration 
- 2 = Inactive ReRegistration
- 3 = Active ReRegistration



It can be seen that the largest retention_d1 have active users, after them, there are new users and at the end, they are inactive users. On average, 25% of new users play again. These are not very good results, because new users don't have a huge preference to play again. 

*Important: Fresh registration may include users who have played before, but this is only their Fresh registration. So we'll explain (drill-down) with "played t11 before" in more detail*

# Imputation NA values

```python
df_train.isna().sum()
```

As we said, the dataset has 0.5% NA values. This is a small percentage, so we can delete this instance. Also, we can imputation NA values with other values. NA values will be changed by the median.

```python
kstest(df_train.loc[df_train['device_memory_size_mb'].isna()==False,'device_memory_size_mb'],'norm')
```

```python
kstest(df_train.loc[df_train['screen_dpi'].isna()==False,'screen_dpi'],'norm')
```

device_memory_size_mb and screen_dpi do not come from a normal distribution (p-value<0.05). Because that, NA values will be changed by the median.

```python
device_memory_size_mb_median=df_train['device_memory_size_mb'].median()

df_train.loc[df_train['device_memory_size_mb'].isna(),'device_memory_size_mb']=device_memory_size_mb_median

screen_dpi_median=df_train['screen_dpi'].median()

df_train.loc[df_train['screen_dpi'].isna(),'screen_dpi']=screen_dpi_median
```

### Distribution function (by_something -  columns/value, ret - filtering of instances that have retention_d1 greater than entered)

```python
def retention_d1_1_by(by_something,ret):
    df_pom=df_train.loc[(df_train["returned"]/df_train["registrations"])>=ret]
    plt.figure(figsize=(7,5))
    df_pom[by_something].value_counts().sort_index().plot(kind='bar')
    plt.xlabel(by_something)
    plt.ylabel('Count')
```

```python
for col in ["registration_type",'played_t11_before',"registration_channel","network_type","device_tier","device_type",'device_manufacturer','week_day_of_registrion']:
    retention_d1_1_by(col,0.8)
```

It can serve to discover the best users.

### The most frequently combination of registration_type, played_t11_before, registration_channel. retention_d1>=0.8

```python
df_train.loc[(df_train["returned"]/df_train["registrations"])>=0.8].groupby(["registration_type",'played_t11_before',"registration_channel"]).size().reset_index().rename(columns={0:'Freq'}).sort_values(by="Freq")
```

The most players with high retention_d1 are players who are active users (registration_type=3) and was targeted by any T11 marketing campaigns (registration_channel=2). After them are players who are active users (registration_type=3) and wasn't targeted by any T11 marketing campaigns (registration_channel=1). After them are the fresh usres (registration_type=1) who was targeted by any T11 marketing campaigns (registration_channel=2).

```python
X_train,y_train=df_train.loc[df_train['date']<=datetime.strptime('2022-08-09', '%Y-%m-%d'),df_train.columns.difference(['date','returned'])],df_train.loc[df_train['date']<=datetime.strptime('2022-08-09', '%Y-%m-%d'),'returned']
X_test,y_test=df_train.loc[df_train['date']>datetime.strptime('2022-08-09', '%Y-%m-%d'),df_train.columns.difference(['date','returned'])],df_train.loc[df_train['date']>datetime.strptime('2022-08-09', '%Y-%m-%d'),'returned']
```

### Results without optimization

```python
models={'GradientBoosting':GradientBoostingRegressor(),
        'RandomForest':RandomForestRegressor(n_jobs=-1),
        'XGBRegressor':XGBRegressor(n_jobs=-1)}

results={}
for name,concrete_model in models.items():
    print(name)
    model=concrete_model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    
    X_test_pom=df_train.loc[df_train['date']>datetime.strptime('2022-08-09', '%Y-%m-%d')]
    X_test_pom['pred']=y_pred

    a=X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['returned']/X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['registrations']

    b=X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['pred']/X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['registrations']

    results[name]=(np.sqrt(mean_squared_error(y_test,y_pred)),np.sqrt(mean_squared_error(np.round_(a.values*100,3),np.round_(b.values*100,3))))
```

```python
#konacno
(pd.DataFrame(results,index=['rmse_returned','rmse_retention_d1']).T).sort_values(by='rmse_retention_d1')
```

### Results with optimization

```python
models={'GradientBoosting':GradientBoostingRegressor(learning_rate=0.1,max_depth=3,n_estimators=300,subsample=1),
        'RandomForest':RandomForestRegressor(max_depth=8, min_samples_leaf=5, n_estimators=50,n_jobs=-1),
        'XGBRegressor':XGBRegressor(learning_rate=0.1,max_depth=3,n_estimators=300,subsample=1,n_jobs=-1,tree_method='hist')}

results={}
for name,concrete_model in models.items():
    print(name)
    model=concrete_model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    
    X_test_pom=df_train.loc[df_train['date']>datetime.strptime('2022-08-09', '%Y-%m-%d')]
    X_test_pom['pred']=y_pred

    a=X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['returned']/X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['registrations']

    b=X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['pred']/X_test_pom.groupby('date').agg({'registrations':'sum','pred':'sum','returned':'sum'})['registrations']

    results[name]=(np.sqrt(mean_squared_error(y_test,y_pred)),np.sqrt(mean_squared_error(np.round_(a.values*100,3),np.round_(b.values*100,3))))
```

```python
#konacno
(pd.DataFrame(results,index=['rmse_returned','rmse_retention_d1']).T).sort_values(by='rmse_retention_d1')
```

It can be seen that by optimizing the parameters reduces rmse_returned. But, with a smaller rmse_returned, rmse_retention_d1 doesn't decrease. Therefore, the non-optimized model RandomForest will be used.

The best results were obtained using the optimized GradientBoosting, if the RMSE for each combination is observed. Since the goal is to make predictions aggregated by days, the best results are given by unoptimized RandomForest. This is why this model will be used.
